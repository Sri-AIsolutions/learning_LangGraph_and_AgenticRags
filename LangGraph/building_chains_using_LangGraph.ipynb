MEssages:
there are different types of messages like AImessage, Human message, System message, tool message

every message has these important components:

1->content---content if the message

2->name---specific name of the author

3->response_metadata

```python
from langchain_core.messages import AIMessage,HumanMessage
#creating custom messages manually---->these basically happen during the chat (but here we r just creating it manually for better understanding)
messages=[AIMessage(content=f"heyy! tell me what do u want to learn today!!!",name="LLMModel")]
messages.append(HumanMessage(content=f"I want to master RAGs and Agentic AI",name="sri"))
messages.append(AIMessage(content=f"Do u want to start with Agent Archietecture or Langchain Basics",name="LLMModel"))

for n in messages:
    n.pretty_print()



```

while building the graph --we make sure messages are appended in its stateschema automatically 

NOTE: these automatic appending part is completely taken care by reducers in langgraphs
